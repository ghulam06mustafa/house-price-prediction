```python
x=df[['number of bedrooms','number of bathrooms','living area','condition of the house','Number of schools nearby']]
y=df[['Price']]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x.shape)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lrmodel=LinearRegression()
```

```python
lrmodel
```

```python
lrmodel.fit(x_train,y_train)
```

```python
y_predict=lrmodel.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
maelr=mean_absolute_error(y_test,y_predict)
mse=mean_squared_error(y_test,y_predict)
r2Score=r2_score(y_test,y_predict)
```

```python
maelr
```

```python
mse
```

```python
r2Score
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=1.0)
```

```python
ridge.fit(x_train,y_train)
```

```python
y_predictRidge=ridge.predict(x_test)
```

```python
import numpy as np
maeridge=mean_absolute_error(y_test,y_predictRidge)
mseridge=np.sqrt(mean_squared_error(y_test,y_predictRidge))
r2scoreRidge=r2_score(y_test,y_predictRidge)
```

```python
maeridge
```

```python
mseridge
```

```python
r2scoreRidge
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
rfy_predict=rf.predict(x_test)
```

```python
rfmae=mean_absolute_error(y_test,rfy_predict)
rfmse=np.sqrt(mean_squared_error(y_test,rfy_predict))
rfr2score=r2_score(y_test,rfy_predict)
```

```python
print(rfmae)
print(rfmse)
print(rfr2score)
```
